```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
###import csv files
mouse_drug_data1= "data/mouse_drug_data.csv"
clinical_trial_data2 = "data/clinicaltrial_data.csv"
```

```python
###Read the csv files with panda library
trial_data = pd.read_csv(mouse_drug_data1)
mouse_data = pd.read_csv(clinical_trial_data2)
```

```python
### Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data=pd.read_csv(mouse_drug_data1)
clinical_data=pd.read_csv(clinical_trial_data2)
mouse_data.columns

### Combine the data into a single dataset form
data_combined=pd.merge(mouse_data, clinical_data, how="left", on=["Mouse ID", "Mouse ID"])

### Display the data table for preview
data_combined.head()
data_combined.columns
data_combined=data_combined[["Mouse ID", "Timepoint", "Tumor Volume (mm3)",
                            "Metastatic Sites", "Drug"]]
data_combined.head()
```

# Tumor Volume During Treatment

```python
###data_complete["Drug"].value_counts()
###data_complete["Timepoint"].value_counts()
tumor_volume=data_combined.loc[:,["Drug", "Timepoint", "Tumor Volume (mm3)"]]


### Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol=tumor_volume.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean().to_frame()

### Convert to DataFrame
mean_df=pd.DataFrame(mean_tumor_vol.loc[:,["Tumor Volume (mm3)"]])
```

```python
###Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sterror_tumor_vol=tumor_volume.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem().to_frame()
###Convert to DataFrame
sterror_df=pd.DataFrame(sterror_tumor_vol.loc[:,["Tumor Volume (mm3)"]])
```

```python
### Minor Data Munging to Re-Format the Data Frames
means=mean_df.unstack(0)
means_df=means["Tumor Volume (mm3)"] #goes down a level
sterror=sterror_df.unstack(0)
sterror_df=sterror["Tumor Volume (mm3)"] #goes down a level
```

```python
###Preview that Reformatting dataset
means_df
sterror_df
```

```python
x_axis = means.index
plt.figure()
```

```python
### Generate the Plot (with Error Bars)
error = sterror["Tumor Volume (mm3)"]["Capomulin"]
c=plt.errorbar(x_axis, means_df["Capomulin"], yerr=error, fmt="o", ls="dotted", color="blue")

error = sterror["Tumor Volume (mm3)"]["Infubinol"]
i=plt.errorbar(x_axis, means_df["Infubinol"], yerr=error, fmt="^", ls="dotted", color="red")

error = sterror["Tumor Volume (mm3)"]["Ketapril"]
k=plt.errorbar(x_axis, means_df["Ketapril"], yerr=error, fmt="s", ls="dotted", color="yellow")

error = sterror["Tumor Volume (mm3)"]["Placebo"]
p=plt.errorbar(x_axis, means_df["Placebo"], yerr=error, fmt="d", ls="dotted", color="green")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(linestyle="-")
plt.legend((c,i,k,p),("Capomulin", "Infubinol", "Ketapril", "Placebo"))
plt.show()

### Save the Figure
plt.savefig("tumor_response.jpg")
```
